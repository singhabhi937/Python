```python
#function for creating a data frame in python
```

```python


def make_df(cols, ind):
    """Quickly make a DataFrame"""
    data = {c: [str(c) + str(i) for i in ind]
            for c in cols}
    return pd.DataFrame(data, ind)

# example DataFrame
make_df('ABC', range(3))

```

```python
#concat function is quite tricky as it has various parameters and how does it concatenate depends on those parameters and the index values
```

```python
# case one same column name but different index
d1=make_df('ABC', [1,2])
d2=make_df('ABC', [4,5])
```

```python
d3=pd.concat([d1,d2])
```

```python
d3 # since we did not pass any extra prameter it was expected to do it like that only 
```

```python
# concatenate using axis 
# when indexes are differnent in that case default is row bind 
```

```python
d3=pd.concat([d1,d2], axis=1) #1 is for 'cols' 
```

```python
d3
```

```python
# when indexws are same what is the default concatenation defa==
d1=make_df('ABC', [1,2])
d2=make_df('CDE', [2,3])
```

```python
d3=pd.concat([d1,d2]) #1 is for 'cols' 
```

```python
d3 #default concatenation is always row bind 
```

```python
d3=pd.concat([d1,d2], verify_integrity=True) #this is for avoiding the concatenation of dupllicate index
```

```python
d3
```

```python
# now how to reset the index 
```

```python
d1=make_df('ABC', [1,2])
d2=make_df('CDE', [2,3])
```

```python
d3=pd.concat([d1,d2],ignore_index=True)
```

```python
d3
```
